```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.image as mpimg
import numpy as np
import matplotlib as mpl
mpl.get_backend()
```

```python
def drawContour(data1,data2,data3,lev):
    '''
    param: data1 :
        X data - np.array
    param: data2 :
        Y data - np.array
    param: data3 :
        Z data - np.meshgrid
    param: levels :
    '''
    
    plt.contour(data1,data2,data3,levels=lev,cmap='hot')
    plt.colorbar()
```

```python
X = np.array([1,2,3,4])
Y = np.array([1,2,3,4,5])
Z = np.ones([len(Y),len(X)])
for i in range(len(Y)):
    for j in range(len(X)):
        Z[i][j] = np.sin((i+1)*(j+2))
print(Z)
```

```python
drawContour(X,Y,Z,0)
```
