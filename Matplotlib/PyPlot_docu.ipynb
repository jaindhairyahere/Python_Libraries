```python
# Property	Value Type
# alpha	float
# animated	[True | False]
# antialiased or aa	[True | False]
# clip_box	a matplotlib.transform.Bbox instance
# clip_on	[True | False]
# clip_path	a Path instance and a Transform instance, a Patch
# color or c	any matplotlib color
# contains	the hit testing function
# dash_capstyle	['butt' | 'round' | 'projecting']
# dash_joinstyle	['miter' | 'round' | 'bevel']
# dashes	sequence of on/off ink in points
# data	(np.array xdata, np.array ydata)
# figure	a matplotlib.figure.Figure instance
# label	any string
# linestyle or ls	[ '-' | '--' | '-.' | ':' | 'steps' | ...]
# linewidth or lw	float value in points
# marker	[ '+' | ',' | '.' | '1' | '2' | '3' | '4' ]
# markeredgecolor or mec	any matplotlib color
# markeredgewidth or mew	float value in points
# markerfacecolor or mfc	any matplotlib color
# markersize or ms	float
# markevery	[ None | integer | (startind, stride) ]
# picker	used in interactive line selection
# pickradius	the line pick selection radius
# solid_capstyle	['butt' | 'round' | 'projecting']
# solid_joinstyle	['miter' | 'round' | 'bevel']
# transform	a matplotlib.transforms.Transform instance
# visible	[True | False]
# xdata	np.array
# ydata	np.array
# zorder	any number
```

```python
fig = plt.figure()
fig, ax= plt.subplots(2,2)
def hello():
    print('Hello World')
ax[0][0].plot([9,16,25,36,49,64,81,100],'r4',aa=False,alpha=0.25)
ax[0][0].plot([9,16,35,46,49,84,91,121],'r4',aa=False,zorder=1,solid_capstyle='round')

ax[0][1].plot([9,16,25,36,49,64,81,100],'b^',animated=True,markevery=2)
ax[1][0].plot([9,16,25,36,49,64,81,100],marker ='+',linewidth=4,linestyle='steps',label='yoyo bitch',contains=hello)
ax[1][1].plot([9,16,25,36,49,64,81,100],mfc='r',linewidth=0,marker='.',markeredgecolor = 'yellow', mew=3,markersize=10)




#animated
#contains
#antialiased
#picker
#pickradius
# solid_capstyle
# solid_joinstyle

# transform	a matplotlib.transforms.Transform instance 
# clip_box	a matplotlib.transform.Bbox instance
# clip_on	[True | False]
# clip_path	a Path instance and a Transform instance, a Patch
# dash_capstyle	['butt' | 'round' | 'projecting']
# dash_joinstyle	['miter' | 'round' | 'bevel']
# dashes	sequence of on/off ink in points
```

```python
#Working with Multiple Figures

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
f(1)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)                 #Currently subplot of 2*1 grid and 1st subplot --> Figure 1
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')         #t1,f(t1) in blue o(s), and t2,f(t2) in black line
plt.subplot(212)                 #Currently subplot of 2*1 grid and 2st subplot --> Figure 1
plt.plot(t1,t1)

plt.figure(2)
plt.plot(t1,np.cos(t1))

plt.figure(2)


#Adding Text to Figure at a point (3,0) on it
plt.text(3,0, u'Hello World')
plt.grid(True)
```
